```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver  
import re
import urllib
import time
```

```python
# @brief 保存图片 # url : 图片url # addr  : 保存的地址
def save_picture(url,addr,name):
    urllib.request.urlretrieve(url,'{}{}.jpg'.format(addr,name))    
    return
```

```python
item_list = pd.read_csv('E:/Downloads/picture/data_information/蕾丝猫套图信息.csv',engine='python')
```

```python
x = item_list['link'].iloc[1:10]
y = x[~x.isin([18257,1,2])]
for i in y:
    print(i)
```

```python
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
dcap = dict(DesiredCapabilities.PHANTOMJS)
dcap["phantomjs.page.settings.userAgent"] = ("Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36")
driver = webdriver.PhantomJS(executable_path="D:/Python/selenium/webdriver/phantomjs/bin/phantomjs.exe", desired_capabilities=dcap)  
```

```python
#path = 'C:\\Users\\Think\\Desktop\\pythondownload\\'
path = 'E:/Downloads/picture/蕾丝猫套图/'
for link in item_list['link'].iloc[5482:]: 
    time.sleep(5)
    j = 1
    for num in range(1,6):
        url1 = 'http://www.lesmao.me/thread-' + str(link) + '-%s-1.html'%str(num)
        print('正在加载图片%s'%(url1))
        try:
            driver.get(url1)
            data = driver.page_source
            soup = BeautifulSoup(data,'lxml')
            lis1 = soup.find('div',attrs={'id':'thread-pic'}).find_all('li')
            for lis in lis1:
                picture = lis.find('img')['src']
                save_picture(picture,path,str(link)+'_'+str(j))
                j += 1
        except:
            print('加载失败')
```
