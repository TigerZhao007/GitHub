```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver  
import re
import urllib
import time
```

```python
url_list = ['http://dx5y.com/html/part/index28.html']+['http://dx5y.com/html/part/index28_%s.html'%(str(num)) for num in range(2,70)]
```

```python
item = {
    'name':'小说名称',
    'date':'更新日期',
    'link':'小说链接',
    'text':'小说内容'
}
item = pd.DataFrame.from_dict(item,orient='index').T
```

```python
browser = webdriver.PhantomJS(executable_path="D:/Python/selenium/webdriver/phantomjs/bin/phantomjs.exe")  
```

```python
item_list = pd.DataFrame()
for url in url_list:
    print('--------------------------------------------------------')
    print('正在加载%s信息'%(url))
    browser.get(url)
    data = browser.page_source
    soup =  BeautifulSoup(data,'lxml')
    lis1 = soup.find('div',attrs={'class':'box list channel'}).find('ul').find_all('li')[5:]
    for lis in lis1:        
        item['link'] =  'http://dx5y.com' + lis.find('a')['href']
        item['date'] = lis.find('span').getText()
        item['name'] = re.findall(r'</span>(.+?)</a>',str(lis))[0]
        try:
            browser.get('http://dx5y.com' + lis.find('a')['href'])
            data = browser.page_source
            soup =  BeautifulSoup(data,'lxml')
            item['text'] = soup.find('div',attrs={'class':'content'}).getText().replace('\n','').replace(' ','')
        except:
            item['text'] = 'NA'
        item_list = item_list.append(item)
```

```python
item_list.to_csv("C:/Users/Think/Desktop/test.csv",index=False,sep=',',encoding='utf_8_sig')
```
