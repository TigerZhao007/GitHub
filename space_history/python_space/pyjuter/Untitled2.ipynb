```python
import requests
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://nj.lianjia.com/ershoufang/'
```

```python
url_list = ['https://nj.lianjia.com/ershoufang/pg%s/' %(str(num)) for num in range(1,20)]
```

```python
content = requests.get(url).content
```

```python
url_list
```

```python
soup = BeautifulSoup(content,'lxml')
```

```python
item = {
    'houseid':'房屋编码',
    'housename':'房屋名字',
    'houseprice':'房屋价格',
}
item = pd.DataFrame.from_dict(item,orient='index').T
```

```python
lis = soup.find('ul',attrs = {'class':'sellListContent'}).find_all('li')
```

```python
item['housename'] = lis[0].find('div',attrs = {'class':'title'}).text.replace('房主自荐','')
```

```python
item['houseid'] = lis[0].find('div',attrs = {'class':'houseInfo'}).text
```

```python
item['houseprice'] = lis[0].find('div',attrs = {'class':'totalPrice'}).text
```

```python
item_list = item
for url in url_list:
    print('--------------',url)
    content = requests.get(url).content
    soup = BeautifulSoup(content,'lxml')
    lis = soup.find('ul',attrs = {'class':'sellListContent'}).find_all('li')
    for li in lis:
        
        try:
            item['housename'] = li.find('div',attrs = {'class':'title'}).text.replace('房主自荐','')
            item['houseid']   = li.find('div',attrs = {'class':'houseInfo'}).text
            item['houseprice'] = li.find('div',attrs = {'class':'totalPrice'}).text
            item_list = item_list.append(item)
        except:
            print('爬取房源失败')    
```
