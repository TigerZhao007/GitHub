```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url = 'https://nj.lianjia.com/ershoufang/pg3/'
```

```python
content = requests.get(url).content
```

```python
soup = BeautifulSoup(content,'lxml')
```

```python
lis = soup.find('ul',attrs = {'class':'sellListContent'}).find_all('li')
```

```python
item = {
    'houseid':'房屋编码',
    'housename':'房屋名字',
    'houseprice':'房屋价格',
}
item = pd.DataFrame.from_dict(item,orient='index').T
```

```python
item['houseid'] = lis[0].find('div',attrs = {'class':'title'}).text
```

```python
item['housename'] = lis[0].find('div',attrs = {'class':'houseInfo'}).text
```

```python
item['houseprice'] =lis[0].find('div',attrs = {'class':'totalPrice'}).text
```

```python
item_list = item
for li in lis:
    try:
        item['houseid'] = li.find('div',attrs = {'class':'title'}).text
        item['housename'] = li.find('div',attrs = {'class':'houseInfo'}).text
        item['houseprice'] =li.find('div',attrs = {'class':'totalPrice'}).text
        item_list = item_list.append(item)
    except:
        print('错误')
```
